```python
#import beatiful soup and requests for webscraping 
import requests 
from bs4 import BeautifulSoup 


#import pandas and numpy for eventual compiliation of data
import pandas as pd 
import numpy as np
```

```python
#Import webpage
# For now i'm only going to focus on 2020 ipos
url = "https://stockanalysis.com/ipos/2020/"
page = requests.get(url)

#If we print page.text --> we can see all html code and objects from the page
print(page.text)
```

```python
#Convert the webpage to a soup object
ipoSoup = BeautifulSoup(page.content, "html.parser")
ipoSoup 
```

```python
# Now we want to pull out the table itself for scraping and ultimate transition 
# First we'll find the Ipo table's id from the element 

findthetable = ipoSoup.find(id="ipo-table")
#print(fin)
tableOrganized = (findthetable.prettify())
ipotable = findthetable
print(ipotable)
```

```python
# Now comes our work with beautiful soup 

# first we attempt find the tickers which come as links 'a' in the table with the class 'bll'
tickers = ipotable.find_all("a", class_ = "bll")

# print out just the  stock ticker abreviation using the .string attribute

for ticker in tickers:
    #print(type(ticker))
    #print(ticker) #we notice the ticker is an bs4 element, specifically an html hyperlink
    
    print(ticker.string)
```

```python
# Now that we have the tickers all at once, lets try to get the date 
# by navigating backwards in the beautifulSoup tree:
# We'll try to create a dictionary with all the relevant ipos 
# we 
ipoDict = {}
counter = 0
for ticker in tickers:
    #print(ticker.string)
    tckr = ticker.string
    counter +=1
    #print(counter)
    #we want the date the stock ipod
    #print(ticker.parent.previous_sibling.string) #So, we go up an element to the td tag, then we go sideways to the left, previous sibling, then we use the .string attribute to only get the string contained withing the element
    date = ticker.parent.previous_sibling.string
    #we want the full name of the stock
    #print(ticker.parent.next_sibling.string)
    fullname = ticker.parent.next_sibling.string
    #we want the price it ipo'd at 
    #print(ticker.parent.next_sibling.next_sibling.string)
    ipoprice = ticker.parent.next_sibling.next_sibling.string
    ipoDict[fullname]=[tckr,date,ipoprice]
ipoDict
```

```python
len(ipoDict)
#There's an issue here, because there is actually 480 ipos in 2020
# this is a problem that occurs because some tckrs are hyperlinks and others arent 
```

```python
# Lets try this by looking at the body
# I can do this by looking at the soup object, and then .(tagname)
# ipotable.tbody gives me body of the table 
# ipotable.tody.contents gives me a list of whats in the body (by row)
# ipotable.tbody.children gives me an iterable of all the children 


# print(ipotable)

type(ipotable)
counter = 0
ipodict = {}
for row in ipotable.tbody.children:
    counter +=1
    #print(row.td) #gives me first td tag in table row, which is the date
    date = row.td.string
    tckr = row.td.next_sibling.string
    fullname = row.td.next_sibling.next_sibling.string
    ipoprice = row.td.next_sibling.next_sibling.next_sibling.string
    ipodict[fullname] = [tckr,date,ipoprice]
len(ipodict) #problem solved - just had to work top down 
ipodict
```

```python
# Put the dictionary into a dataframe 
ipoDF = pd.DataFrame.from_dict(ipodict, orient = 'index', dtype = 'string').reset_index()
#Rename the columns
ipoDF = ipoDF.rename(columns = {'index':'company',0:'ticker',1:'ipodate',2:'ipoprice'})
ipoDF
```

```python
# We need to change the date format 
# To do that we will use the dateutil library, which make converting dates very easy 
import dateutil
from dateutil.parser import parse
sampledate = "Jan 17, 2020"
mydate = parse(sampledate)
#print(str(mydate)[0:10])

# We create a function we can .apply to a column (series) in our ipo dataframe
def changeDateFormat(date):
    parsedDate = parse(date)
    newDate = str(parsedDate)[0:10]
    return newDate

print(changeDateFormat(sampledate))
```

```python
#Apply date formating changes and export to csv
ipoDF['ipodate'] = ipoDF['ipodate'].apply(changeDateFormat)
ipoDF
ipoDF.to_csv("IPO2020")
```
