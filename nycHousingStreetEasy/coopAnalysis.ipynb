```python
from dis import dis


def sixmonthAVG(hoodsOfInterest): 
    discountDF = pd.read_csv('rentData/discountShare_TwoBd.csv')
    priceDF = pd.read_csv('rentData/medianAskingRent_TwoBd.csv')
    inventoryDF = pd.read_csv('rentData/rentalInventory_TwoBd.csv')

    
    
    # Initilalize DataFrame
    result = pd.DataFrame()

    # Inventory avg (1st column)
    inventoryDF = inventoryDF[inventoryDF['areaName'].isin(hoodsOfInterest)].set_index('areaName')
    result['AVG Inventory'] = inventoryDF.loc[:,'2022-02':'2022-07'].mean(axis=1)

    # Price avg
    priceDF = priceDF[priceDF['areaName'].isin(hoodsOfInterest)].set_index('areaName')
    result['AVG Price'] = priceDF.loc[:,'2022-02':'2022-07'].mean(axis=1)

    # Discount Avg
    discountDF = discountDF[discountDF['areaName'].isin(hoodsOfInterest)].set_index('areaName')
    result['AVG Price Reduction'] = discountDF.loc[:,'2022-02':'2022-07'].mean(axis=1)

    return result

sixmonthAVG(myinterest)
```
